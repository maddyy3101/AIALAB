```python
#2(i) BFS
from collections import deque

def bfs(graph, start):
    q = deque([start]); visited = {start}; order=[]
    while q:
        u = q.popleft(); order.append(u)
        for v in graph.get(u, []):
            if v not in visited:
                visited.add(v); q.append(v)
    return order

if __name__=="__main__":
    g = {
        'A': ['B','C'],
        'B': ['D','E'],
        'C': ['F'],
        'D': [], 'E': ['F'], 'F': []
    }
    print("BFS order:", bfs(g,'A'))
```

```python
#2(ii) DFS

def dfs(graph, start):
    visited=set(); order=[]
    def go(u):
        visited.add(u); order.append(u)
        for v in graph.get(u,[]):
            if v not in visited:
                go(v)
    go(start); return order

if __name__=="__main__":
    g = {'A':['B','C'],'B':['D','E'],'C':['F'],'D':[],'E':['F'],'F':[]}
    print("DFS order:", dfs(g,'A'))
```

```python
#4(i && ii) tic tac toe
def print_board(board):
    print(f"{board[0]} | {board[1]} | {board[2]}")
    print("--+---+--")
    print(f"{board[3]} | {board[4]} | {board[5]}")
    print("--+---+--")
    print(f"{board[6]} | {board[7]} | {board[8]}")
    print()


def show_positions():
    print("Positions are numbered 1 to 9 as below:")
    print("1 | 2 | 3")
    print("--+---+--")
    print("4 | 5 | 6")
    print("--+---+--")
    print("7 | 8 | 9")
    print()


def is_valid_move(board, move):
    return move.isdigit() and 1 <= int(move) <= 9 and board[int(move) - 1] == ' '


def check_winner(board, player):
    win_conditions = [
        (0, 1, 2), (3, 4, 5), (6, 7, 8),       # Rows
        (0, 3, 6), (1, 4, 7), (2, 5, 8),       # Columns
        (0, 4, 8), (2, 4, 6)                   # Diagonals
    ]
    for i, j, k in win_conditions:
        if board[i] == board[j] == board[k] == player:
            return True
    return False


def is_draw(board):
    return ' ' not in board


def make_ai_move(board):
    for i in range(9):
        if board[i] == ' ':
            board[i] = 'O'
            return i + 1


def main():
    board = [' '] * 9
    print("Welcome to Tic-Tac-Toe!")
    print("You are X, AI is O")
    show_positions()
    print_board(board)

    while True:
        move = input("Enter your move (1-9): ")

        if not is_valid_move(board, move):
            print("Please enter a valid move.\n")
            continue

        move = int(move) - 1
        board[move] = 'X'
        print_board(board)

        if check_winner(board, 'X'):
            print("ðŸŽ‰ You win!")
            break

        if is_draw(board):
            print("It's a draw!")
            break

        ai_pos = make_ai_move(board)
        print(f"AI placed O at position {ai_pos}")
        print_board(board)

        if check_winner(board, 'O'):
            print("ðŸ¤– AI wins!")
            break

        if is_draw(board):
            print("It's a draw!")
            break


if __name__ == "__main__":
    main()
```

```python
#5(i) Build a Propositional Logic Reasoning System that evaluates logical expressions using connectives (AND, OR, NOT, â†’, â†”).
def implies(p, q):
    return (not p) or q     # p â†’ q â‰¡ Â¬p âˆ¨ q

def iff(p, q):
    return (p and q) or (not p and not q)   # p â†”ï¸ q
propositions = {
    'A': True,
    'B': False,
    'C': True
}

def evaluate_expressions():
    print("\n--- Propositional Logic Evaluation ---")

    result1 = propositions['A'] and not propositions['C']
    print("A âˆ§ Â¬C =", result1)

    result2 = (propositions['A'] or propositions['B']) and propositions['C']
    print("(A âˆ¨ B) âˆ§ C =", result2)

    result3 = not propositions['B'] or propositions['A']
    print("Â¬B âˆ¨ A =", result3)

    result4 = implies(propositions['A'], propositions['C'])
    print("A â†’ C =", result4)

    result5 = iff(propositions['B'], propositions['C'])
    print("B â†”ï¸ C =", result5)

evaluate_expressions()
```

```python
#5(ii)Propositional Logic Evaluator with User Input
import re

def eval_prop(expr, vals):
    e = expr.replace("âˆ§"," and ").replace("âˆ¨"," or ")
    e = e.replace("Â¬"," not ").replace("->"," or not ")
    for k,v in vals.items():
        e = e.replace(k, str(v))
    return eval(e)

def main():
    expr = input("Enter expression: ")
    props = sorted(set(re.findall(r"[a-zA-Z]", expr)))
    vals = {p: (input(f"{p} (T/F): ").upper()=="T") for p in props}
    print("Result:", eval_prop(expr, vals))

main()
```

```python
#7(i && ii) Design a rule-based Expert System for Medical Diagnosis using forward chaining in Python.
def main():
    rules = [
        (["fever","cough"], "flu"),
        (["cough","sneeze"], "cold"),
        (["headache","nausea"], "migraine")
    ]

    sym = input("Enter symptoms: ").split()
    res=[]
    for r, dis in rules:
        if set(r).issubset(sym):
            res.append(dis)

    print("Possible diseases:", res if res else "None")

main()
```

```python
#a* (i)
import heapq

def main():
    n = int(input("No. of nodes: "))
    graph = {}
    for _ in range(n):
        a = input("Node: ")
        e = input("Edges (format: node1,weight node2,weight ...): ").split()
        graph[a] = [(x.split(",")[0], int(x.split(",")[1])) for x in e]

    start = input("Start: ")
    goal = input("Goal: ")

    h = {}  
    for _ in range(n):
        a = input(f"Heuristic for {input('Node name again: ')}: ")
        h[a] = int(input())

    pq = [(0 + h[start], 0, start, None)]
    parent = {}
    visited = set()

    while pq:
        f, g, u, par = heapq.heappop(pq)
        if u in visited: continue
        visited.add(u)
        parent[u] = par
        if u == goal:
            path = []
            while u:
                path.append(u)
                u = parent[u]
            print("A* Path:", path[::-1], "Cost:", g)
            return
        for v, w in graph[u]:
            if v not in visited:
                heapq.heappush(pq, (g + w + h[v], g + w, v, u))

main()
```
